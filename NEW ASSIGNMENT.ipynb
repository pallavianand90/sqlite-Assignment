```python
# Reading the txt file and Converting txt file into list of elements
def list_conversion(filename):
    try:
        lg.info("Trying to open a text file")
        a_file = open(filename, "r",encoding='utf-8')
        lg.info("Successfully opened!!!")
        list_of_list1 = []
        for line in a_file:
            stripped_line = line.strip()
            #line_list = stripped_line.split()
            list_of_list1.append(stripped_line)
        return list_of_list1
            #a_file.close()
    except Exception as e:
        lg.exception("Error occured: "+ str(e))
        lg.error("Something went Wrong")
        lg.warning("Fix the bug otherwise code will not work!!!!")
```

```python
list1 =list_conversion("vocab.enron.txt")
list2 =list_conversion("vocab.kos.txt")
list3 =list_conversion("vocab.nips.txt")
list4 =list_conversion("vocab.nytimes.txt")
list5 =list_conversion("vocab.pubmed.txt")
```

```python
# 1) Try to find out a count of each and every word in a respective file return a list of tuple with word and its respective count 

from collections import Counter
def count_element(a):
    try:
        lg.info("Trying to count the elements...")
        col_count = Counter(a).most_common()
        lg.info("Successufully done!!")
        return col_count
    
    except Exception as e:
        lg.exception("Error occured: "+ str(e))
        lg.error("Something went Wrong")
        lg.warning("Fix the bug otherwise code will not work!!!!")

    
```

```python
#2) try to perform a reduce operation to get a count of all the word starting with same alphabet.

from pandas import DataFrame,Series
def count_starting_letter(a):
    try:
        lg.info("Creating series...")
        s = Series([word[0] for word in a])
        lg.info("Created!!")
        z = s.value_counts()
        lg.info("Creating frame")
        y =z.to_frame()
        lg.info("Done")
        lg.info("Creating Records")
        records = y.to_records(index=True)
        lg.info("Done")
        result = list(records)
        lg.info("Done successfully")
        print(result)
    except Exception as e:
        lg.exception("Error occured: "+ str(e))
        lg.error("Something went Wrong")
        lg.warning("Fix the bug otherwise code will not work!!!!")
     
```

```python
#3) Try to filter out all the words from dataset .

import re
try:
    lg.info("Converting list to string..")
    listToStr = ' '.join(map(str, list5))
    lg.info("Done!!!")
    lg.info("Spliting the string with some selective delimiter...")
    sep = re.split("[.' '-3' '8]",listToStr)
    lg.info("Done!!!")
except Exception as e:
    lg.exception("Error occured: "+ str(e))
    lg.error("Something went Wrong")
    lg.warning("Fix the bug otherwise code will not work!!!!")
     
l = []
lg.info("Selecting words mixed strings..")   
for i in sep:
    if len(i)>=2 and i.isalpha()==True:
        l.append(i)
        lg.info("done!!")
print(l)
```
